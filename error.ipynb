```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
LinearInterpolation = []
Kriging = []
DeepKriging = []
BaseModel = []
```

```python
LinearInterpolation = []
Kriging = []
DeepKriging = []
BaseModel = []

for x in range(2,22,2):
    print(x)
    with open(f'newplots/main/wholeMap_x-{x}_y-{x}_RandomSampling/error.txt') as f :
        
        lines = f.readlines()
        
        for i in range(10):
            
            if 'Linear Interpolation' in lines[i]:
                
                LinearInterpolation.append(float(lines[i][lines[i].find(':')+1:-2]))
            
            if ('Kriging' in lines[i] and not 'Deep ' in lines[i]):
                
                Kriging.append(float(lines[i][lines[i].find(':')+1:-2]))
            
            if 'Deep Kriging' in lines[i]:
                
                DeepKriging.append(float(lines[i][lines[i].find(':')+1:-2]))
                
            if 'Base Model' in lines[i]:
                
                BaseModel.append(float(lines[i][lines[i].find(':')+1:-2]))
            
                
                f.close()
        
```

```python
maes1 = pd.DataFrame({
              'Kriging' : Kriging,
              'Linear' : LinearInterpolation
              })
```

```python
maes = pd.DataFrame({
              'DeepKriging' : DeepKriging,
              'BaseModel' : BaseModel
              })
```

```python
maes.plot(figsize=(12,6), grid=True, title='MAE over different sampling distances for uniformly resampled data')
plt.ylabel('MAE')
plt.xlabel('sampling distance in cm') 
plt.savefig('MaeUniformampling.png')
```
